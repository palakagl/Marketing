```python
df = pd.read_csv('order_products__train_detailed.csv')
print('\nNumber of rows and columns in the data set: ',df.shape)
print('')

#Lets look into top few rows and columns in the dataset
df.head(15)
```

```python
combined_data = df[['order_id','product_name']]
combined_data.head(20)
```

```python
onehot = combined_data.pivot_table(index='order_id', columns='product_name', aggfunc=len, fill_value=0)
onehot = onehot>0
```

```python
from mlxtend.frequent_patterns import association_rules, apriori

# compute frequent items using the Apriori algorithm
frequent_itemsets = apriori(onehot, min_support = 0.000001, use_colnames=True)

# compute all association rules for frequent_itemsets
rules = association_rules(frequent_itemsets)
rules.head(20)
```

```python
rules.sort_values('lift', ascending=True).head(5)
```
